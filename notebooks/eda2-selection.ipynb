### Selection decision

Given the heterogeneity of the data through the years, the data used in the current analysis and in the simulation project will only take into account data related to the "Year One" section. Furthermore, data for "Year One" is the most complete of the three sets, containing not only trip and station data, but also status and weather.

```python
data = {
    "trip": selector.dfs["2015_trip_data"],
    "station": selector.dfs["2015_station_data"],
    "status": selector.dfs["2015_status_data"],
    "weather": selector.dfs["2015_weather_data"]
}
```

```python
selector.serialize(data=data, serial_path="data/selected.pickle")
```
